```python
#delete the country codes -99
world_cities = world_cities[world_cities.iso2 != '-99']
```

```python
##check again how many unique countries
unique_countries = world_cities['iso2'].unique()
print len(unique_countries)
print world_cities.shape
```

```python
# handle_strings raised error for codes - float object object has no attribute 'upper'
# change to str

world_cities['iso2'] = world_cities['iso2'].astype(str)
world_cities['city_ascii'] = world_cities['city_ascii'].astype(str)
```

```python
world_cities['Clean_City'] = world_cities['city_ascii'].apply(handle_strings)
world_cities['Clean_Code'] = world_cities['iso2'].apply(handle_strings)
```

##Prep for Fuzzy-Wuzzy

```python
correct_cities = world_cities['Clean_City'].tolist()
len(correct_cities)
```

```python
def correct_city(cityname):
    if cityname in correct_cities: 
        return cityname, 100

    new_name, score = process.extractOne(cityname, correct_cities)
    if score < 81:
        return cityname, score
    else:
        return new_name, score
```

###Unicode Decode Error

Unicode gave me problem on the reference list. First identify where exactly is the problem then handle:

```python
correct_cities_success = []
for city in correct_cities:
    success = [c for c in city if ord(c) < 128]
    success =''.join(success)
    correct_cities_success.append(success)            
```

```python
# print list(set(correct_cities_success) - set(correct_cities))
print list(set(correct_cities) - set(correct_cities_success))
```
