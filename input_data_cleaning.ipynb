#Cleaning the input_data 

####Below cleans the cities input data: drops NAs, normalize strings etc. 

In total 2 769 590 cities - it used to be 2 794 888 before dropping NA City names

```python
import pandas as pd
input_data = pd.read_csv('Data/Input_Data.txt', sep='|', index_col=False, 
                 names=['City', 'Country Code', 'X', 'Y'])
```

in some lines there are 4 fields. that's why I added 2 other columns: X,Y 

```python
input_data = input_data.drop(input_data.index[[0]])
```

```python
input_data.head()
```

```python
len(input_data)
```

```python
x = input_data.sort('X')
y = input_data.sort('Y')
```

##Note
on data frame x - City and Country Code are str but X and Y were float. In order to merge, converted to str

```python
print type(x['City'][1])
print type(x['Country Code'][1])
print type(x['X'][1])
print type(x['Y'][1])
```

```python
x[['X', 'Y']] = x[['X', 'Y']].astype(str)
```

```python
print type(x['City'][1])
print type(x['Country Code'][1])
print type(x['X'][1])
print type(x['Y'][1])
```

```python
import numpy as np
x = x.replace('nan', '')
x.tail()
```

```python
x.head()
```

```python
x['City_New'] = x[['City', 'X']].apply(lambda a: ''.join(a), axis=1)
```

```python
x.sort('X', ascending = False)
```

```python
df = x[['City_New','Country Code','Y']]
```
